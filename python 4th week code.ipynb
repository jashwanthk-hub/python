```python
print("=====list examples====")
print("apple", "cherry", "orange")  # duplicate allowed

# define the list properly
fruits = ["apple", "banana", "cherry", "orange"]

print("original list:", fruits)

# remove element
fruits.remove("banana")

# access by index
print("first fruit:", fruits[0])

# list comprehension
upper_fruits = [fruit.upper() for fruit in fruits]
print("uppercase fruits:", upper_fruits)
```

```python
print ("\n===tuple example ===")
coordinates=(10.5,20.3,10.5)
print("Tuple:",coordinates)
print("first coordinates:",coordinates [-1])
#tuple unpacking 
x,y,z=coordinates
print (f"unpacked:x={x},y={y},z={z}")
```
